```python
monthly = pd.read_csv('data/monthly_sales.csv')
display(monthly.head(2))
print(monthly.shape)
```

### product sale distribution among different areas

```python
all_time_total_sales = monthly[[i for i in monthly.columns if '-' in i]].sum(axis = 0).rename('total_sale_by_area_product').to_frame()
```

```python
all_time_total_sales.plot(kind = 'bar')
```

Every area has similar product poopularity. <br>
Note: we do not know about whether our product sales "equally" well among all regions because the data is not normalized to the population of each region.

#### first analyze by food type

```python
monthly['HM_total'] = monthly[[i for i in monthly.columns if 'HM' in i]].sum(axis = 1)
monthly['CF_total'] = monthly[[i for i in monthly.columns if 'CF' in i]].sum(axis = 1)
monthly['FF_total'] = monthly[[i for i in monthly.columns if 'FF' in i]].sum(axis = 1)
```

```python
total_per_food = monthly[[i for i in monthly.columns if 'total' in i]]
total_per_food.plot(kind = 'line')
plt.plot([33, 33], [total_per_food.min().min(), total_per_food.max().max()])
```

Every food type experiences the largest drop in Oct-18. BK has its new vegan burger out at the same month.<br>
Coincidence? I think not.

```python
total_per_food.sum().rename('total sale by product').plot(kind = 'pie')
```

This chart digresses, but it's also fun to think about the popularity of products. Hamburgers > chicken > fish. Typical 'Muricans.

#### then by area type

```python
monthly['NE_area_total'] = monthly[[i for i in monthly.columns if '-NE' in i]].sum(axis = 1)
monthly['SW_area_total'] = monthly[[i for i in monthly.columns if '-SW' in i]].sum(axis = 1)
monthly['NW_area_total'] = monthly[[i for i in monthly.columns if '-NW' in i]].sum(axis = 1)
monthly['SE_area_total'] = monthly[[i for i in monthly.columns if '-SE' in i]].sum(axis = 1)
monthly['C_area_total'] = monthly[[i for i in monthly.columns if '-C' in i]].sum(axis = 1)
```

```python
total_per_area = monthly[[i for i in monthly.columns if 'area' in i]]
```

```python
total_per_area.plot(kind = 'line')
plt.plot([33,33], [total_per_area.min().min(), total_per_area.max().max()])
```

As it turns out that the sales by area is a bit redundant. Probably will not include it in the final viz report.

```python
total = total_per_food.sum(axis = 1).rename('total sales').to_frame()
total.plot(kind = 'line')
plt.plot([33,33], [total.min(), total.max()])
```

```python
monthly['Month, Year'].loc[33]
```

### Conclusion:
Given that hamburger > chicken fillet > fish fillet, and that all areas sale approximately the same amount of product, every area and every product experiences a huge drop in sale at oct-18.
If we share a zero-sum commercial environment with BK, which I believe is a likely case, then BK's new burger did in fact draw people away for a while and that includes the **non-vegan** ones, implying that people tried it out more likely out of curiosity.

Daily trends on the other hand, is less cryptic than the annual trends. There is a very obvious periodicity presented in the sales trend for fish fillet across all regions.

Singling out trends by region we can see the periodic trends for fish fillet sales.
